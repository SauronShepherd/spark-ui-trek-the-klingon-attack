# 🧪 Lab 2: The Command Log Audit (Jobs Tab Telemetry & Action Verification)

In this laboratory session, we track how specific PySpark operations translate into physical job rows, timelines, and progress metrics inside the Spark UI.

**Objective:** Master the transition from abstract code rows to physical job executions and isolate driver planning gaps from active worker compute blocks.

### Step 1: Boot the Staging Ground Environment
We initialize our localized standalone Spark session and assign an explicit application title to anchor our top-level dashboard telemetry.

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
import time

spark = (
    SparkSession.builder
                .appName("spark-ui-jobs-tab-command-log-lab-02")
                .master("local[*]")
                .config("spark.ui.port", "4040")
                .getOrCreate()
)

print(f'Active Engine Version: {spark.version}')
print(f"🛰️ Actual Spark UI: {spark.sparkContext.uiWebUrl}")
```

### Step 2: The Ferengi Trap (Eager Evaluation Audit)
We stage a basic range transformation. We will intentionally insert multiple diagnostic lookup hooks to witness how intermediate development queries force redundant job generation cycles inside the tracker.

```python
spark.sparkContext.setJobDescription('💸 Ferengi Trap: Debug Count Hook')
base_data = spark.range(0, 5000000).withColumn('sector_key', F.col('id') % 50)

# Action 1: Triggers a Spark job (often Job 0 in a fresh session) to evaluate row volumes
check_count = base_data.count()

spark.sparkContext.setJobDescription('💸 Ferengi Trap: Cellular Layout Render')
# Action 2: Triggers another job to render visual elements to the console
base_data.show(2)

print(f'📊 Ingestion checkpoints evaluated: {check_count} rows registered.')
```

### Step 3: Isolating Driver-Side Lag Footprints
We halt execution on the driver thread while Spark is doing no distributed work; executors or local task slots are idle. This generates a massive, distinct vacuum inside the chronological event timeline tracker.

```python
print('⚠️ Freezing processing loops to simulate heavy non-Spark driver overhead...')
time.sleep(15)
print('⚡ Driver thread resumed. Preparing follow-up production execution tracks.')
```

### Step 4: Execute the Production Target Materialization
We execute our actual destination write command, spawning the final write-side job or jobs, depending on the execution plan and environment. We lock the runtime active link to allow complete profiling of the compiled timeline logs.

```python
spark.sparkContext.setJobDescription('🚀 Production Target: Parquet Export Runway')
export_path = '/tmp/spark_ui_jobs_log_demo_export'

base_data.write.mode('overwrite').parquet(export_path)

print('\n⏳ Process complete. Open port 4040 and audit the gaps between your active job rows now!')
time.sleep(30)
spark.stop()
```
