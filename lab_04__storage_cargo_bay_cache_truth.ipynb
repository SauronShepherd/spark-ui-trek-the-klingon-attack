# 🧪 Lab 4: The Cargo Bay Interrogation (Storage Tab Matrix & Eviction Controls)

In this laboratory session, we audit Spark's cache allocation engine to observe how lazy declarations react to active materialization commands and investigate memory eviction under resource constraints.

**Objective:** Verify the lazy behavior of `.cache()`, identify partial partition materialization footprints, and track memory-to-disk spill behaviors natively within port `:4040`.

### Step 1: Initialize the Telemetry Viewscreen
We spin up our isolated standalone Spark session and enforce a local storage buffer memory profile to simulate execution pressure targets.

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
import time

spark = (
    SparkSession.builder
                .config("spark.driver.memory", "2g")
                .getOrCreate()
)

print(f'Active Core Engine: {spark.version}')
print('🛰️ Telemetry connection open. Monitoring storage state targets.')
```

### Step 2: The Lazy Cache Blueprint Request
We generate a dense numerical telemetry dataframe and invoke a standard `.cache()` statement. Because Spark operations are strictly lazy, no jobs are spawned, and your storage manifest remains completely blank.

**Mission Directive:** Run the cell below, then immediately inspect the Storage tab at port 4040. You will see that absolutely zero objects are registered.

```python
dense_telemetry = (
    spark.range(0, 12000000)
    .withColumn('sector_id', (F.col('id') % 250).cast('int'))
    .withColumn('checksum', F.sha2(F.col('id').cast('string'), 256))
)

# Requesting cache space (Lazy contract declaration)
staged_cache_df = dense_telemetry.cache()
print('🛡️ Cache request submitted. Storage tab remains unmaterialized.')
```

### Step 3: Materialize the Cargo Manifest (The Action Command)
We execute a terminal count action over our dataset. This forces Spark to evaluate partitions, load block instances into memory, and populate our Storage summary fields with active telemetry metrics.

```python
spark.sparkContext.setJobDescription('🧊 Materializing In-Memory Cargo Blocks')

start_time = time.time()
total_rows = staged_cache_df.count()
end_time = time.time()

print(f'🎯 Cache materialized successfully: {total_rows} entries packed.')
print(f'⏱️ Operational materialization clock: {end_time - start_time:.2f} seconds.')
print('⏳ Navigate to the Storage tab now! Verify the 100% fraction cached status.')
```

### Step 4: The Unpersist Clean Room Protocol
We execute a manual cleanup sequence to drop the allocated memory footprints, proving how cell lifecycle management updates our active system profiles.

```python
print('🧹 Evicting staged cache lines to clear the cargo bay...')
staged_cache_df.unpersist()

print('\n⏳ Subspace sweep complete. Verify that port 4040 storage views have cleared cleanly!')
time.sleep(30)
spark.stop()
```
